```python
# Visualize the dataset
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('iris') 

sns.set(style="whitegrid")
```

```python
# Scatter plot between two numerical features
```

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='petal_length', y='petal_width', hue='species')
plt.title('Scatter Plot: Petal Length vs Sepal Width')
plt.show()
```

```python
# Plot histograms for all numeric columns
```

```python
df.hist(figsize=(10, 8), bins=20, edgecolor='black')
plt.suptitle('Histograms of All Numeric Features')
plt.tight_layout()
plt.show()
```

```python
# Box plots for detecting outliers in each feature
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df.drop(columns='species'), orient='h', palette="Set2")
plt.title('Box Plots of Features')
plt.show()
```
